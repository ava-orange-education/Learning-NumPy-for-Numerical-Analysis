## **Outlier Analysis**

### **Z-score method**

```python
#Z-score method for RM (Average Number of Rooms)
mean_rm = np.mean(boston_df['RM'])
std_rm = np.std(boston_df['RM'])

# Z-score formula: (value - mean) / standard deviation
rm_z_scores = (boston_df['RM'] - mean_rm) / std_rm

# Identify outliers where Z-score > 3 or Z-score < -3
rm_outliers_z = boston_df[(rm_z_scores > 3) | (rm_z_scores < -3)]

# Count the number of outliers found using the Z-score method
print("Number of outliers using Z-scores:", rm_outliers_z.shape[0])

# Display RM values identified as outliers using Z-scores
rm_outliers_z[['RM']]
```

### **IQR method**

```python
# Example#1 IQR method for RM
Q1 = boston_df['RM'].quantile(0.25)
Q3 = boston_df['RM'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier threshold using 1.5 * IQR rule
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers using IQR method
rm_outliers_iqr = boston_df[(boston_df['RM'] < lower_bound) | (boston_df['RM'] > upper_bound)]

# Display RM values identified as outliers using IQR method
print("Number of outliers for variable RM using IQR:", rm_outliers_iqr.shape[0])

# Display first 5 RM values identified as outliers using IQR method
print("First 5 RM values identified as outliers using IQR:", rm_outliers_iqr['RM'][:5])


#Example#2 IQR for CRIM
Q1 = boston_df['CRIM'].quantile(0.25)
Q3 = boston_df['CRIM'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier threshold using 1.5 * IQR rule
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers using IQR method
crim_outliers_iqr = boston_df[(boston_df['CRIM'] < lower_bound) | (boston_df['CRIM'] > upper_bound)]

# Display CRIM values identified as outliers using IQR method
print("\nNumber of outliers for variable CRIM using IQR:", crim_outliers_iqr.shape[0])

# Display first 5 CRIM values identified as outliers using IQR method
print("First 5 CRIM values identified as outliers using IQR:", crim_outliers_iqr['CRIM'][:5])
```

### **Outliers Analysis using boxplot**

```python
# Customize boxplot properties
boxprops = dict(linewidth=2, color='blue', facecolor='lightblue')
medianprops = dict(linewidth=2, color='red')
whiskerprops = dict(linewidth=2, color='black')
capprops = dict(linewidth=2, color='black')

# Get all the column names from the dataset
columns = boston_df.columns

# Determine the number of rows needed
n_cols = 5
n_rows = len(columns) // n_cols + (len(columns) % n_cols > 0)

# Set the number of subplots (4 columns per row) and adjust the layout
fig, axs = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 3))

# Loop through each column and create a boxplot
for i, col in enumerate(columns):
    row = i // n_cols
    col_idx = i % n_cols
    axs[row, col_idx].boxplot(boston_df[col], vert=False, patch_artist=True, showmeans=True,
                              boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops,
                              capprops=capprops, meanline=True, meanprops=dict(color='green', linewidth=2))
    axs[row, col_idx].set_title(f'IQR Boxplot for {col}', fontsize=12)
    axs[row, col_idx].set_xlabel('Values', fontsize=10)
    axs[row, col_idx].grid(True)


for i in range(len(columns), n_rows * n_cols):
    fig.delaxes(axs.flatten()[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

### **Univariate Analysis**


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Univariate analysis of MEDV (Median value of homes in $1000s)
plt.figure(figsize=(8, 6))
sns.histplot(boston_df['MEDV'], kde=True, color='blue')
plt.title('Distribution of House Prices (MEDV)')
plt.xlabel('House Price ($1000s)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Remove CHAS and RAD as they are categorical variables
cols = [col for col in boston_df.columns if col not in ['CHAS', 'RAD']]
plot_per_col = 4

# Create subplots: Adjusting rows and columns for numeric columns
fig, axes = plt.subplots(3, plot_per_col, figsize=(10, 8))  # Adjusted figsize for better visibility

# Loop through each variable and create the histograms
for i, col in enumerate(cols):
    row = i // plot_per_col  # Determine row index
    col_pos = i % plot_per_col  # Determine column index
    sns.histplot(boston_df[col], kde=True, ax=axes[row, col_pos], color='blue')
    axes[row, col_pos].set_title(f'Distribution of {col}')
    axes[row, col_pos].set_xlabel(col)
    axes[row, col_pos].set_ylabel('Frequency')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
